```python
import pandas as pd
import io

df = pd.read_csv('most_subscribed_youtube_channels.csv')
display(df)
```

```python
df = df.replace(',', '')
display(df)
```

```python
import plotly.express as px
fig = px.histogram(df, x='subscribers', title='Subscriber Count')
fig.show()
```

```python
import plotly.express as px

fig = px.pie(df, values='subscribers', names='category', title='YouTube Categories')
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns




# Set the style using Seaborn
sns.set(style="whitegrid")

# Create a pie chart
plt.figure(figsize=(8, 8))  # Adjust the figure size as needed
plt.pie(df['subscribers'], labels=df['category'], autopct='%1.1f%%', startangle=140)
plt.title('YouTube Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace this with your actual DataFrame)


# Convert 'subscribers' column to numeric by removing commas
df['subscribers'] = df['subscribers'].str.replace(',', '').astype(float)

# Create the pie chart
plt.figure(figsize=(8, 8))  # Adjust the figure size as needed
plt.pie(df['subscribers'], labels=df['category'], autopct='%1.1f%%', startangle=140)

# Set the title
plt.title('YouTube Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Show the pie chart
plt.show()
```
