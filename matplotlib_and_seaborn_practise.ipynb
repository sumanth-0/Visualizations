```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle='dashdot',linewidth=2)
```

```python
plt.plot(x,y,color='g',linestyle='dashdot',linewidth=2)
j=np.arange(1,14)
x1=j*2
y1=j*3
print(y1)
plt.plot(x1,y1,color='g',linestyle=':',linewidth=2)
plt.xlabel('ok')
plt.ylabel('jok')
plt.title("tryyying")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='g',linestyle='dashdot',linewidth=2)
j=np.arange(1,14)
x1=j*2
y1=j*3
print(y1)
plt.subplot(1,2,2)
plt.plot(x1,y1,color='g',linestyle=':',linewidth=2)
plt.xlabel('ok')
plt.ylabel('jok')
plt.title("tryyying")
plt.grid(True)
plt.legend()
plt.show()
```
